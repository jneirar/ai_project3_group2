```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KDTree
import gdown
from PIL import Image, ImageOps
from scipy.spatial import distance
from scipy.stats import multivariate_normal
import seaborn as sns
import os
import json
from math import log
import os
from sklearn.neighbors import BallTree,DistanceMetric
from IPython.display import display
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
data = np.array(load_iris().data)
labels_pre = np.array(load_iris().target)
labels = []
for i in labels_pre:
    label_arr = np.zeros(3)
    label_arr[i] = 1
    labels.append(label_arr)
labels = np.array(labels)

#Normalizaci√≥n:
sc = MinMaxScaler()
sc.fit(data)
data = sc.transform(data)

n = data.shape[0]
m = data.shape[1]
indexes = np.arange(n)
x_train, x_oth, y_train, y_oth = train_test_split(data, labels, test_size=0.2)
x_test, x_validation, y_test, y_validation = train_test_split(x_oth, y_oth, test_size=0.5)

print(x_train)
print(x_test.shape)
print(x_validation.shape)
print(y_train.shape)
print(y_test.shape)
print(y_validation.shape)
```

```python
obj = dict()
obj["Train"] = {"Classes": y_train.tolist(), "Images": x_train.tolist()}
obj["Test"] = {"Classes": y_test.tolist(), "Images": x_test.tolist()}
obj["Validation"] = {"Classes": y_validation.tolist(), "Images": x_validation.tolist()}
with open("data.json", "w") as write_file:
    json.dump(obj, write_file)
```
